```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv("/Users/saniyapareshkirkire/Desktop/DSCI_554/Homework6/a6-Sanaaa2000/Data/SYB65_145_202209_Land.csv",skiprows=1)
df2 = pd.read_csv("/Users/saniyapareshkirkire/Desktop/DSCI_554/Homework6/a6-Sanaaa2000/Data/SYB65_315_202209_Water_and_Sanitation_Services.csv",skiprows=1)
df3 = pd.read_csv("/Users/saniyapareshkirkire/Desktop/DSCI_554/Homework6/a6-Sanaaa2000/Data/SYB65_313_202209_Threatened_Species.csv",skiprows=1)
df4 = pd.read_csv("/Users/saniyapareshkirkire/Desktop/DSCI_554/Homework6/a6-Sanaaa2000/Data/SYB65_310_202209_Carbon_Dioxide_Emission_Estimates.csv",skiprows=1)
```

```python
df4.head()
```

```python
df1 = df1[df1['Region/Country/Area'] != 1]
df2 = df2[df2['Region/Country/Area'] != 1]
df3 = df3[df3['Region/Country/Area'] != 1]
df4 = df4[df4['Region/Country/Area'] != 1]
```

```python
df1 = df1.drop(columns=['Region/Country/Area','Footnotes','Source'])
df2 = df2.drop(columns=['Region/Country/Area','Footnotes','Source'])
df3 = df3.drop(columns=['Region/Country/Area','Footnotes','Source'])
df4 = df4.drop(columns=['Region/Country/Area','Footnotes','Source'])
```

```python
df1 = df1.reset_index()
df1 = df1.iloc[:, 1:]
df1 = df1.rename(columns={'Unnamed: 1': 'Countries_Region'})
df2 = df2.reset_index()
df2 = df2.iloc[:, 1:]
df2 = df2.rename(columns={'Unnamed: 1': 'Countries_Region'})
df3 = df3.reset_index()
df3 = df3.iloc[:, 1:]
df3 = df3.rename(columns={'Unnamed: 1': 'Countries_Region'})
df4 = df4.reset_index()
df4 = df4.iloc[:, 1:]
df4 = df4.rename(columns={'Unnamed: 1': 'Countries_Region'})
```

```python
pivoted_df1 = df1.pivot(index=['Countries_Region', 'Year'], columns='Series', values='Value').reset_index()
pivoted_df1 = pivoted_df1.reset_index(drop=True)
pivoted_df1.head()
```

```python
pivoted_df2 = df2.pivot(index=['Countries_Region', 'Year'], columns='Series', values='Value').reset_index()
pivoted_df2 = pivoted_df2.reset_index(drop=True)
pivoted_df2.head()
```

```python
pivoted_df3 = df3.pivot(index=['Countries_Region', 'Year'], columns='Series', values='Value').reset_index()
pivoted_df3 = pivoted_df3.reset_index(drop=True)
pivoted_df3.head()
```

```python
pivoted_df4 = df4.pivot(index=['Countries_Region', 'Year'], columns='Series', values='Value').reset_index()
pivoted_df4 = pivoted_df4.reset_index(drop=True)
pivoted_df4.head()
```

```python
desired_countries = ['Australia', 'United States of America', 'China', 'India', 'Norway']
land_df = pivoted_df1[pivoted_df1['Countries_Region'].isin(desired_countries)]
water_df = pivoted_df2[pivoted_df2['Countries_Region'].isin(desired_countries)]
threat_df = pivoted_df3[pivoted_df3['Countries_Region'].isin(desired_countries)]
co2_df = pivoted_df4[pivoted_df4['Countries_Region'].isin(desired_countries)]
```

```python
land_df.to_csv('/Users/saniyapareshkirkire/Desktop/DSCI_554/Homework6/a6-Sanaaa2000/Data/land.csv', index=False)
water_df.to_csv('/Users/saniyapareshkirkire/Desktop/DSCI_554/Homework6/a6-Sanaaa2000/Data/water.csv', index=False)
threat_df.to_csv('/Users/saniyapareshkirkire/Desktop/DSCI_554/Homework6/a6-Sanaaa2000/Data/threat.csv', index=False)
co2_df.to_csv('/Users/saniyapareshkirkire/Desktop/DSCI_554/Homework6/a6-Sanaaa2000/Data/co2.csv', index=False)
```
